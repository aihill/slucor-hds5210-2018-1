# Timing Algorithms



```python
def pivot1(H, L, n):
  output = []


  for i in range(n, len(L)):
    out = L[:n]


    out.append(H[i])
    out.append(L[i])
    output.append(out)


  return output
```

```python
def pivot2(H, L, n):
  output = []

  copies = len(L) - n
  for i in range(copies):
    out = L[:n]

    copy_index = n + i
    for c in range(len(L)):
        if c == copy_index:
            out.append(H[c])
            out.append(L[c])
            
    output.append(out)

  return output
```

```python
import time

H = ['Head '+str(n) for n in range(1,1000)]
L = ['Row '+str(n) for n in range(1,1000)]
n = 3

# Time #1
start1 = time.perf_counter()
a = pivot1(H,L,n)
end1 = time.perf_counter()
print("pivot1 : " + str((end1 - start1) * 1000))

# Time #2
start2 = time.perf_counter()
a = pivot2(H,L,n)
end2 = time.perf_counter()
print("pivot2 : " + str((end2 - start2) * 1000))
```
